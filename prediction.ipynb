```python
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
ann_model = load_model("model.h5")
```

```python
df = pd.DataFrame([input_data.values()], columns = input_data.keys())
df
```

```python
## convert the gender
with open("labelencoder.pkl", "rb") as f:
    labelencoder = pickle.load(f)

with open("ohe.pkl", "rb") as f:
    ohe = pickle.load(f)

with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)
```

```python
df["Gender"] = labelencoder.transform(df["Gender"])
df
```

```python
df_geography = ohe.transform([df["Geography"]])
df_geography
```

```python
df_geography.toarray()
```

```python
df_geography = pd.DataFrame(df_geography.toarray(), columns = ohe.get_feature_names_out())
df_geography
```

```python
df = pd.concat([df.drop(["Geography"], axis = 1), df_geography], axis = 1)
df
```

```python
scaled_X = scaler.transform(df)
scaled_X
```

```python
value = ann_model.predict(x=scaled_X)
```

```python
if value > 0.5:
    print("He is going to exit")
else:
    print("He is not going to exit")
```
